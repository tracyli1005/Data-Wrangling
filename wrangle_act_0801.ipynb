#### 评分分子错误
dog_rate表格：部分评分分子有误，如用户835246439529840000 的狗评分应为13，被误写为960

##### 定义
检查评分分子大于20的行，并基于text文本语义，修正相应的分子

##### 代码

```python
#提取出分子有小数点，分母为非0的10的倍数#
rating=dograte_clean.text.str.extract('((?:\d+\.)?\d+)\/([1-9]+[0]+)',expand=True)
```

```python
#提取出的Dataframe 两列分别命名为分子和分母#
rating.columns=['rating_numerator','rating_denominator']
```

```python
#用新提取出的数据替换原有数据，因为可能存在小数，数据改为float类型#
dograte_clean['rating_numerator']=rating['rating_numerator'].astype(float)
dograte_clean['rating_denominator']=rating['rating_denominator'].astype(float)
```

```python
dograte_clean.rating_denominator.value_counts()
```

```python
#使用向量运算，把分母为10的倍数的项目，调整为分母为10，且分子做同比调整#
dograte_clean['rating_numerator']=dograte_clean['rating_numerator']/dograte_clean['rating_denominator'] *10
dograte_clean['rating_denominator']=10
```

```python
#检查提取了2个score的情况#
pd.set_option('max_colwidth',200)
dograte_clean['scores']=dograte_clean.text.str.findall('((?:\d+\.)?\d+)\/([1-9]+[0]+)')
dograte_clean['scores_counts']=dograte_clean['scores'].apply(lambda x: len(set(x)))
dograte_clean[dograte_clean['scores_counts']>1][['text','scores','scores_counts']]
```

```python
#针对有2个狗评分的项目取均值#
df=[766,1007,1222,1359,1465,1508,1525,1538,1795,1832,1897,1901,2064,2113,2177,2216,2263,2272]
```

```python
for x in df:
    a=float(dograte_clean.scores[x][0][0])
    b=float(dograte_clean.scores[x][1][0])
    dograte_clean.loc[x,'rating_numerator']=(a+b)/2
    print(dograte_clean.rating_numerator[x])
```

```python
#将有语义中有2个评分的狗狗评分进行校正#
dograte_clean.loc[1165,'rating_numerator']=13
dograte_clean.loc[1202,'rating_numerator']=11
dograte_clean.loc[1970,'rating_numerator']=8
dograte_clean.loc[2010,'rating_numerator']=10
```

#### 质量度错误
dograte表格：
- 部分狗对应多个地位

##### 定义
把dograte表格中，找出stage信息重复的记录，并只保留第1条
针对一只狗狗同时对应几个地位的情况，把stage标注为'multiple'

##### 代码

```python
#把需要合并的列和id单独抓出来作为df_melt,清洗后与原数据集合并#
df_melt=dograte_clean[['tweet_id','doggo','floofer','pupper','puppo']].copy()
df_melt = pd.melt(df_melt,id_vars=['tweet_id'],value_name='stage')

df_melt.head() 
```

```python
#把所有stage 为none的数据删除#
df_melt.drop('variable',axis=1,inplace=True)
df_melt=df_melt[df_melt['stage']!='None']
df_melt.head()
```

```python
#筛选出tweet_id仍重复的项，修改其stage为multiple#
df_melt.loc[df_melt['tweet_id'].duplicated(keep=False),'stage']='multiple'
```

```python
#删除重复项，然后与原数据集做合并#
df_melt.drop_duplicates(inplace=True)
df_melt.stage.value_counts()
```

```python
dograte_clean=dograte_clean.merge(df_melt,how='left',on='tweet_id')
```

##### 测试

```python
dograte_clean.stage.value_counts()
```

```python
twitter_archive_master.groupby('stage').rating_numerator.mean().sort_values().plot(kind='bar')
plt.title('Average Rate by Stage')
plt.xlabel('Stage')
plt.ylabel('Avg. Rate')
```

Doggo(大型犬或较年长的狗)平均得分最高，为12.02分，接下来是Floofer（多毛）和Puppo(青少年期)的狗狗，平均得分为12分，而得分最低的是pupper(狗崽)，说明大型犬或较年长的狗狗更受评委青睐一些。
